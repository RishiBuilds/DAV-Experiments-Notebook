**Experiment Title**: To implement Hypothesis Testing using Python Programming

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
print("Dataset Preview:")
print(df.head())
```

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nDataset Description:")
print(df.describe())
```

```python
math_scores = df['math score']
```

```python
plt.figure(figsize=(8,5))
sns.histplot(math_scores, kde=True)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
population_mean = 65
alpha = 0.05
```

```python
t_statistic, p_value = stats.ttest_1samp(math_scores, population_mean)
```

```python
print("\nOne Sample T-Test Results")
print("-------------------------")
print(f"T-Statistic : {t_statistic}")
print(f"P-Value : {p_value}")
```

```python
print("\nHypothesis Decision:")
if p_value <= alpha:
    print("Reject the Null Hypothesis (H0)")
    print("There is a significant difference between the sample mean and population mean.")
else:
    print("Accept the Null Hypothesis (H0)")
    print("There is no significant difference between the sample mean and population mean.")
```
